```python
import math

opt = int(input("Press 1 for encryption and 0 for decryption.\n"))
```

```python
#@title
if opt == 1:
    KEY = int(input("Enter number of columns\n"))

    plain_text = input("Enter plaintext\n")
    len_of_column = int(math.ceil(len(plain_text) / KEY))
    
    num_spaces = KEY - len(plain_text) % KEY
    for i in range(num_spaces):
        plain_text +=' '
    
    encoded = ''

    for i in range(KEY):
        for j in range(len_of_column):
            encoded += plain_text[i+j*KEY]
        encoded += '$'
    
    print("Encrypted:", encoded, "\n")

elif opt == 0:
    encoded = input("Enter encrypted text\n")
    plain_text = ''
    len__of_column = 0
    for i in encoded:
        if i == "$":
            break
        len__of_column += 1

    for i in range(len__of_column):
        incr = 0
        while i + incr*(len__of_column + 1) < len(encoded):
            plain_text += encoded[i + incr*(len__of_column + 1)]
            incr += 1

    print("Decrypted:", plain_text, "\n")
```

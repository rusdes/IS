```python
import random

p = int(input('Enter Prime number: '))
```

```python
for i in range(2,p):
  li=[]
  alpha = i
  found = False

  for j in range(0,p-1):
    if ((i**j)%p) in li:
      break
    li.append((i**j)%p)
    if j == p-2:
      found = True
  if found:
    break
```

```python
senderPrivateKey = random.randint(2,p)
recieverPrivateKey = random.randint(2,p)

senderPublicKey = pow(alpha, senderPrivateKey, p)
recieverPublicKey = pow(alpha, recieverPrivateKey, p)

senderSecretKey = pow(recieverPublicKey, senderPrivateKey, p)
recieverSecretKey = pow(senderPublicKey, recieverPrivateKey, p)
```

```python
print("\nSender's Side:\nP = ",p,"\nAlpha = ",alpha,"\nPrivate Key = ",senderPrivateKey,"\nPublic key = ",senderPublicKey,"\nSecret Key = ",senderSecretKey)
print("\nReceivers's Side: \nP = ",p,"\nAlpha = ",alpha,"\nPrivate Key = ",recieverPrivateKey,"\nPublic key = ",recieverPublicKey,"\nSecret Key = ",recieverSecretKey)
```

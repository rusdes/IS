```python
#@title
opt = int(input("Press 1 for encryption and 0 for decryption.\n"))
KEY = int(input("Enter key\n"))
```

```python
#@title
if opt == 1:
    plain_text = input("Enter plaintext\n")
    encoded = ''
    for i in plain_text:
        if i == " ":
            encoded += '$'
        else:
            c = ord(i) + KEY - ord('a')
            c %= 26
            c += ord('a')
            encoded += chr(c)
    
    print("Encrypted:", encoded, "\n")

elif opt == 0:
    encoded = input("Enter encrypted text\n")
    plain_text = ''
    for i in encoded:
        if i == '$':
            plain_text += ' '
        else:
            c = ord(i) - KEY - ord('a')
            c %= 26
            c += ord('a')
            plain_text += chr(c)
    
    print("Decrypted:", plain_text, "\n")
```

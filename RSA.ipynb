```python
import random
max_PrimLength = 100
```

```python
def find_mod_inv(a,b):
  d = 2
  while True:
    if (a*d)%b == 1:
      break
    d += 1
  return d
```

```python
# calculates the modular inverse from e and phi
def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)
```

```python
# calculates the gcd of two ints
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

# checks if a number is a prime
def is_prime(num):
    if num == 2:
        return True
    if num < 2 or num % 2 == 0:
        return False
    for n in range(3, int(num**0.5)+2, 2):
        if num % n == 0:
            return False
    return True
```

```python
def generateRandomPrim():
    while(1):
        ranPrime = random.randint(0,max_PrimLength)
        if is_prime(ranPrime):
            return ranPrime
```

```python
def generate_keyPairs():
    p = generateRandomPrim()
    q = generateRandomPrim()
    
    n = p*q
    print("n =",n)
    phi = (p-1) * (q-1) 
    print("phi =", phi)

    '''choose e coprime to n and 1 > e > phi'''    
    e = random.randint(1, phi)
    g = gcd(e,phi)
    while g != 1:
        e = random.randint(1, phi)
        g = gcd(e, phi)
        
    print("e =",e," ","phi=",phi)
    # d = modular inverse of e and phi
    d = egcd(e, phi)[1]
    
    # make sure d is positive
    d = d % phi
    if(d < 0):
        d += phi
        
    return ((e,n),(d,n))
```

```python
def decrypt(ctext,private_key):
    try:
        key,n = private_key
        text = [chr(pow(char,key,n)) for char in ctext]
        return "".join(text)
    except TypeError as e:
        print(e)

def encrypt(text,public_key):
    key,n = public_key
    ctext = [pow(ord(char),key,n) for char in text]
    return ctext
```

```python
if __name__ == '__main__':
    public_key, private_key = generate_keyPairs() 
    print("Public: ",public_key)
    print("Private: ",private_key)
    
    opt = int(input("Press 1 for encrytion:\n"))
    if opt == 1:
        text = input("Enter text to encrypt:\n")
        ctext = encrypt(text, public_key)
        print("encrypted =",ctext)
        opt = int(input("Press 0 if you want to decrypt this:\n"))
        if opt == 0:
            plaintext = decrypt(ctext, private_key)
            print("decrypted =",plaintext)
```
